```python
import sys
def marks_validate(marks):
    try:
        if marks > 100 or marks < 0:
            raise Exception("ERROR : YOu can not have these marks")
        else:
            print("Got it")
    except:
        print(sys.exc_info())
    finally:
        print("You got it")
```

```python
def fibo(n):
    if n <= 1:
        print(n)
    a = 0
    b = 1
    print(a)
    print(b)
    for i in range(n):
        c = a+b
        a = b
        b = c
        print(c)
```

```python
fibo(100)
```

```python
def fact(n):
    if n <= 1:
        print(n)
    
    fact = 1
    for i in range(1,n+1):
        fact = fact * i

    print(fact)
```
